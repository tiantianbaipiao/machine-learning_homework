```python
import numpy as np #导入numpy 矩阵运算
from matplotlib import pyplot as plt #导入绘图模块 pyplot
from sklearn.model_selection import train_test_split #从sklearn中导入train_test_split函数，功能是从样本中随机的按比例选取训练集与测试集
```

```python
# loading data
data = np.loadtxt('data1.txt',delimiter=',') #加载数据集 delimiter分隔符 np.load_npy 读取npy 读取excel pd.read_excel csv pd.read_csv
print(type(data))
print(data.shape) #打印多少行，多少列
num_feature = data.shape[1] - 1 #输入数据的通道数即特征的数量 特征数量有多长
print(num_feature)
data = data.astype('float32')# 转换为双精度数据类型
```

```python
# data normalization 数据规范化
#data_norm = data.copy()
#maximum = np.max(data_norm,axis=0,keepdims=True) # axis=0代表列 axis=1代表行 keepdims=True结果保持其原来维数 找出列中最大的元素
#print(maximum)
#minimun = np.min(data_norm,axis=0,keepdims=True) #找出列中最小的元素
#data_norm = (data_norm - minimun)/(maximum - minimun) #归一化
#print(data_norm)
```

```python
# train val split 划分数据集
data_train, data_test = train_test_split(data, test_size=0.3, random_state=42) #test_size 划分训练集与测试集 random_state=随机数的种子
# data normalization 归一化
data_norm = data.copy()
maximum = np.max(data_norm,axis=0,keepdims=True) # axis=0代表列 axis=1代表行 keepdims=True结果保持其原来维数 找出列中最大的元素
#print(maximum)
minimun = np.min(data_norm,axis=0,keepdims=True) #找出列中最小的元素

data_train =  (data_train - minimun)/(maximum - minimun) 
#print(data_train)


data_test = (data_test - minimun)/(maximum - minimun) #归一化


#print(data_norm)
X_train = data_train[:, :2] #包括所有自变量，这些变量将用于训练模型 切片 打前两列出来
#print(X_train)
X_train = np.concatenate((X_train, np.ones((X_train.shape[0],1))), axis=1)#拼接数组
print(X_train)
y_train = data_train[:, 2]
print(y_train)
X_test = data_test[:, :2]
X_test = np.concatenate((X_test, np.ones((X_test.shape[0],1))), axis=1)
y_test = data_test[:, 2]
```

```python
# model init
w = np.random.rand(num_feature+1,1)#通过本函数可以返回一个或一组服从标准正态分布的随机样本值。
print(w)
# gradient descent 梯度下降
def L2_loss(y_pred,y):
    return np.mean(np.square(y_pred-y))#loss 函数

iterations = 10000 #迭代次数
lr = 0.1 #学习率
```

```python
log = []
for i in range(iterations):
    y_pred = np.matmul(X_train, w)# 预测值  广播机制 矩阵乘法
    term = lr*np.mean((y_pred-y_train.reshape(-1,1))*X_train, axis=0).reshape(-1,1)#梯度下降
    w -= term
    print(w)
    loss = L2_loss(y_pred,y_train)
    print('iter:{},loss:{}'.format(i,loss))
    log.append([i,loss])
```

```python
# normal eqution 正规方程 套公式
term = np.matmul(X_train.T, X_train)
term_inv = np.linalg.inv(term)# 求逆
w2 = np.matmul(np.matmul(term_inv,X_train.T),y_train.reshape(-1,1))
print(w2)
print(w)
```

```python
# loss curve visualization
log = np.array(log)
plt.plot(log[:,0],log[:,1])
```

```python
# train val split`Êï∞ÊçÆÂàíÂàÜ
data_train, data_test = train_test_split(data, test_size=0.3, random_state=42)

data_norm = data.copy
maximum = np.max(data,axis = 0,keepdims = True)
minimum = np.min(data,axis = 0,keepdims = True)

data_train = (data_train-minimum)/(maximum-minimum)
data_test = (data_test-minimum)/(maximum-minimum)

X_train = data_train[:, :2]
X_train = np.concatenate((X_train, np.ones((X_train.shape[0],1))), axis=1)#??? ÊØèË°åÂä†1 ÊòØ‰∏∫‰∫ÜÂπ≤‰ªÄ‰πà? ÂÅáËÆæÊâÄÊúâËÆ≠ÁªÉÈõÜÊ†∑Êú¨‰∏≠ÁöÑÊ†áÁ≠æÈÉΩ‰∏∫1
y_train = data_train[:, 2].reshape(-1,1)# Ë°®Á§∫Ôºà‰ªªÊÑèË°åÔºå1ÂàóÔºâ
#print(y_train)
X_test = data_test[:, :2]
X_test = np.concatenate((X_test, np.ones((X_test.shape[0],1))), axis=1)
y_test = data_test[:, 2].reshape(-1,1)
```

```python
# model init
w = np.zeros((num_feature+1,1))#3Ë°å‰∏ÄÂàóÁöÑÁü©Èòµ ÂÖÉÁ¥†ÂÖ®‰∏∫0
lambdaw=0.00001
def cross_entropy_loss(y_pred,y):
    return -np.mean(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))#np.log‰ª•10‰∏∫Â∫ï Ëá™ÁÑ∂ÂØπÊï∞  ‰ª£‰ª∑ÂáΩÊï∞Ôºö‚àíùë¶log‚Å°‚ÑéùúÉ(ùë•)‚àí(1‚àíùë¶)log‚Å°(1‚àí‚ÑéùúÉ(ùë•))

def L2_loss(y_pred,y,w):
    return -(np.mean(y*np.log(y_pred)+(1-y)*np.log(1-y_pred)))+np.mean(w*lambdaw)#‰ª£‰ª∑ÂáΩÊï∞Ôºö‚àíùë¶log‚Å°‚ÑéùúÉ(ùë•)‚àí(1‚àíùë¶)log‚Å°(1‚àí‚ÑéùúÉ(ùë•))+lamd/m*ùúÉj

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

iterations = 10000
lr = 0.4
```

```python
# Ê¢ØÂ∫¶ËÆ°ÁÆóÔºàL2Ê≠£ÂàôÂåñÔºâ
logL2 = []
for i in range(iterations):
    y_pred = sigmoid(np.matmul(X_train, w))#‰∏§‰∏™Áü©ÈòµÁõ∏‰πò
    g = lr*(np.mean((y_pred-y_train)*X_train, axis=0).reshape(-1,1) + 2 * w * lambdaw )
    w -= g
    print(w)
    loss = L2_loss(y_pred,y_train,w)
    print('iter:{},loss:{}'.format(i,loss))
    logL2.append([i,loss])
    y_pred_test = sigmoid(np.matmul(X_test, w))
    loss = L2_loss(y_pred_test,y_test,w)
#     print(logL2)
```

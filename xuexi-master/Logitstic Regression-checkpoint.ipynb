```python
# train val split`数据划分
data_train, data_test = train_test_split(data, test_size=0.3, random_state=42)

data_norm = data.copy
maximum = np.max(data,axis = 0,keepdims = True)
minimum = np.min(data,axis = 0,keepdims = True)

data_train = (data_train-minimum)/(maximum-minimum)
data_test = (data_test-minimum)/(maximum-minimum)

X_train = data_train[:, :2]
X_train = np.concatenate((X_train, np.ones((X_train.shape[0],1))), axis=1)#??? 每行加1 是为了干什么? 假设所有训练集样本中的标签都为1
y_train = data_train[:, 2].reshape(-1,1)# 表示（任意行，1列）
#print(y_train)
X_test = data_test[:, :2]
X_test = np.concatenate((X_test, np.ones((X_test.shape[0],1))), axis=1)
y_test = data_test[:, 2].reshape(-1,1)
```

```python
# model init
w = np.zeros((num_feature+1,1))#3行一列的矩阵 元素全为0
lambdaw=0.00001
def cross_entropy_loss(y_pred,y):
    return -np.mean(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))#np.log以10为底 自然对数  代价函数：−𝑦log⁡ℎ𝜃(𝑥)−(1−𝑦)log⁡(1−ℎ𝜃(𝑥))

def L2_loss(y_pred,y,w):
    return -(np.mean(y*np.log(y_pred)+(1-y)*np.log(1-y_pred)))+np.mean(w*lambdaw)#代价函数：−𝑦log⁡ℎ𝜃(𝑥)−(1−𝑦)log⁡(1−ℎ𝜃(𝑥))+lamd/m*𝜃j

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

iterations = 10000
lr = 0.4
```

```python
# 梯度计算（L2正则化）
logL2 = []
for i in range(iterations):
    y_pred = sigmoid(np.matmul(X_train, w))#两个矩阵相乘
    g = lr*(np.mean((y_pred-y_train)*X_train, axis=0).reshape(-1,1) + 2 * w * lambdaw )
    w -= g
    print(w)
    loss = L2_loss(y_pred,y_train,w)
    print('iter:{},loss:{}'.format(i,loss))
    logL2.append([i,loss])
    y_pred_test = sigmoid(np.matmul(X_test, w))
    loss = L2_loss(y_pred_test,y_test,w)
#     print(logL2)
```

```python
# L2 loss curve visualization L2损失函数可视化
logL2 = np.array(logL2)
plt.plot(logL2[:,0],logL2[:,1])
```
